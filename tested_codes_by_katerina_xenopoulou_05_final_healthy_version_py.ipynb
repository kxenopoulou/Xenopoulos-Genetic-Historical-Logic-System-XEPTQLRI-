```python
# ============================================================================
# HEALTHY THERMAL CALCULATOR - XENOPARADOX CORRECTED VERSION
# ============================================================================
# ŒëœÖœÑœåœÇ ŒµŒØŒΩŒ±Œπ Œø Œ•ŒìŒôŒóŒ£ Œ∫œéŒ¥ŒπŒ∫Œ±œÇ ŒºŒµœÑŒ¨ œÑŒπœÇ Œ¥ŒπŒøœÅŒ∏œéœÉŒµŒπœÇ ŒûŒµŒΩœåœÄŒøœÖŒªŒøœÖ
# ŒìŒπŒ± œÑœÅŒ≠ŒæŒπŒºŒø œÉœÑŒø Google Colab
# ============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Œ°œçŒ∏ŒºŒπœÉŒ∑ ŒøœÄœÑŒπŒ∫ŒøœÄŒøŒØŒ∑œÉŒ∑œÇ
plt.style.use('default')
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['font.size'] = 11

print("‚úÖ Œ•ŒìŒôŒóŒ£ ŒòŒïŒ°ŒúŒôŒöŒüŒ£ ŒöŒ©ŒîŒôŒöŒëŒ£ - ŒàŒ∫Œ¥ŒøœÉŒ∑ ŒûŒµŒΩœåœÄŒøœÖŒªŒøœÖ")
print("="*60)

# ============================================================================
# 1. ŒüŒô ŒîŒôŒüŒ°ŒòŒ©Œ£ŒïŒôŒ£ ŒûŒïŒùŒüŒ†ŒüŒ•ŒõŒüŒ• - ŒöŒõŒëŒ£Œó ŒëŒ£Œ¶ŒëŒõŒïŒôŒëŒ£
# ============================================================================

class XenoparadoxSafetyController:
    """ŒïŒªŒµŒ≥Œ∫œÑŒÆœÇ Œ±œÉœÜŒ±ŒªŒµŒØŒ±œÇ Œ≤Œ¨œÉŒµŒπ Œ±ŒΩŒ¨ŒªœÖœÉŒ∑œÇ ŒûŒµŒΩœåœÄŒøœÖŒªŒøœÖ"""

    def __init__(self):
        self.history = []
        self.alerts = []
        self.safety_engaged = False

    def monitor_state(self, A, not_A, result_value):
        """Œ†Œ±œÅŒ±Œ∫ŒøŒªŒøœçŒ∏Œ∑œÉŒ∑ Œ∫Œ±œÑŒ¨œÉœÑŒ±œÉŒ∑œÇ Œ∫Œ±Œπ ŒµŒΩŒµœÅŒ≥ŒøœÄŒøŒØŒ∑œÉŒ∑ Œ±œÉœÜŒ±ŒªŒπœÉœÑŒπŒ∫œéŒΩ"""

        # Œ•œÄŒøŒªŒøŒ≥ŒπœÉŒºœåœÇ Œ¥ŒπŒ±ŒªŒµŒ∫œÑŒπŒ∫ŒÆœÇ Œ≠ŒΩœÑŒ±œÉŒ∑œÇ
        tension = abs(A * not_A)

        # Œ•œÄŒøŒªŒøŒ≥ŒπœÉŒºœåœÇ XEPTQLRI
        xeptqlri = self._calculate_xeptqlri(A, not_A, tension)

        # ŒöŒ±œÑŒ±Œ≥œÅŒ±œÜŒÆ
        state = {
            'A': A,
            'not_A': not_A,
            'tension': tension,
            'xeptqlri': xeptqlri,
            'result': result_value,
            'timestamp': len(self.history)
        }
        self.history.append(state)

        # ŒïŒõŒïŒìŒßŒüŒô ŒëŒ£Œ¶ŒëŒõŒïŒôŒëŒ£ ŒûŒïŒùŒüŒ†ŒüŒ•ŒõŒüŒ•
        self._perform_safety_checks(state)

        return state

    def _calculate_xeptqlri(self, A, not_A, tension):
        """Œ•œÄŒøŒªŒøŒ≥ŒπœÉŒºœåœÇ XEPTQLRI (ŒîŒπŒøœÅŒ∏œâŒºŒ≠ŒΩŒ∑ Œ≠Œ∫Œ¥ŒøœÉŒ∑)"""
        # ŒöŒ±ŒΩŒøŒΩŒπŒ∫ŒøœÄŒøŒØŒ∑œÉŒ∑ œÑŒπŒºœéŒΩ œÄœÅœéœÑŒ±
        norm_A = np.tanh(A)
        norm_not_A = np.tanh(not_A)

        # ŒîŒπŒøœÅŒ∏œâŒºŒ≠ŒΩŒøœÇ œÑœçœÄŒøœÇ
        dialectical_tension = abs(norm_A * norm_not_A)
        paradox_factor = self._calculate_safe_paradox_factor(norm_A, norm_not_A)
        historical_trend = 1.0  # Œ£œÑŒ±Œ∏ŒµœÅœå Œ≥ŒπŒ± Œ±œÄŒªœåœÑŒ∑œÑŒ±

        # ŒåœÅŒπŒø aufhebung: 0.65 (Œ≤Œ≠ŒªœÑŒπœÉœÑŒø)
        aufhebung_threshold = 0.65

        xeptqlri = (dialectical_tension * historical_trend * paradox_factor) / aufhebung_threshold

        # Œ†ŒµœÅŒπŒøœÅŒπœÉŒºœåœÇ œÉŒµ ŒªŒøŒ≥ŒπŒ∫Œ¨ œåœÅŒπŒ±
        return min(max(xeptqlri, 0), 3.0)

    def _calculate_safe_paradox_factor(self, A, not_A):
        """ŒëœÉœÜŒ±ŒªŒÆœÇ œÖœÄŒøŒªŒøŒ≥ŒπœÉŒºœåœÇ œÄŒ±œÅŒ±Œ≥œåŒΩœÑŒ± œÄŒ±œÅŒ±Œ¥œåŒæŒøœÖ"""
        # ŒëœÄŒøœÜœÖŒ≥ŒÆ œÖœÄŒµœÅŒ≤ŒøŒªŒπŒ∫œéŒΩ œÑŒπŒºœéŒΩ
        safe_A = max(-0.9, min(0.9, A))
        safe_not_A = max(-0.9, min(0.9, not_A))

        # ŒíŒµŒªœÑŒπœâŒºŒ≠ŒΩŒøœÇ œÑœçœÄŒøœÇ
        extreme_A = 1 if abs(safe_A) > 0.7 else 0
        extreme_not_A = 1 if abs(safe_not_A) > 0.7 else 0

        # ŒúŒµŒπœâŒºŒ≠ŒΩŒ∑ ŒµœÄŒØŒ¥œÅŒ±œÉŒ∑
        paradox_factor = (extreme_A * extreme_not_A) * 0.8

        return min(paradox_factor, 1.0)

    def _perform_safety_checks(self, state):
        """ŒïŒªŒ≠Œ≥œáŒøœÖœÇ Œ±œÉœÜŒ±ŒªŒµŒØŒ±œÇ Œ≤Œ¨œÉŒµŒπ Œ∫œÅŒπœÑŒ∑œÅŒØœâŒΩ ŒûŒµŒΩœåœÄŒøœÖŒªŒøœÖ"""

        # ŒöŒ°ŒôŒ§ŒóŒ°ŒôŒü 1: Œ•œÄŒµœÅŒ≤ŒøŒªŒπŒ∫ŒÆ Œ≠ŒΩœÑŒ±œÉŒ∑ (> 0.7)
        if state['tension'] > 0.7:
            self.alerts.append({
                'type': 'HIGH_TENSION',
                'message': f'ŒîŒπŒ±ŒªŒµŒ∫œÑŒπŒ∫ŒÆ Œ≠ŒΩœÑŒ±œÉŒ∑ {state["tension"]:.3f} > 0.7',
                'action': 'ŒúŒµŒØœâœÉŒ∑ œÄŒ±œÅŒ±ŒºŒ≠œÑœÅœâŒΩ Œ±ŒΩœÑŒØŒ∏ŒµœÉŒ∑œÇ'
            })
            self.safety_engaged = True

        # ŒöŒ°ŒôŒ§ŒóŒ°ŒôŒü 2: ŒßŒ±ŒºŒ∑ŒªŒÆ Œ≠ŒΩœÑŒ±œÉŒ∑ ŒºŒµ Œ±Œ∫œÅŒ±ŒØŒµœÇ œÑŒπŒºŒ≠œÇ (< 0.3 & |A|>0.8 & |¬¨A|>0.8)
        elif (state['tension'] < 0.3 and
              abs(state['A']) > 0.8 and
              abs(state['not_A']) > 0.8):
            self.alerts.append({
                'type': 'PARADOX_RISK',
                'message': 'ŒöŒØŒΩŒ¥œÖŒΩŒøœÇ œÄŒ±œÅŒ±Œ¥ŒøŒæŒøŒªŒøŒ≥ŒπŒ∫ŒÆœÇ œÖœÄŒ≠œÅŒ≤Œ±œÉŒ∑œÇ',
                'action': 'ŒëœçŒæŒ∑œÉŒ∑ Œ≥ŒΩŒÆœÉŒπŒ±œÇ Œ¥ŒπŒ±ŒªŒµŒ∫œÑŒπŒ∫ŒÆœÇ Œ≠ŒΩœÑŒ±œÉŒ∑œÇ'
            })

        # ŒöŒ°ŒôŒ§ŒóŒ°ŒôŒü 3: Œ•œàŒ∑ŒªœåœÇ XEPTQLRI (> 1.0)
        if state['xeptqlri'] > 1.0:
            self.alerts.append({
                'type': 'HIGH_XEPTQLRI',
                'message': f'XEPTQLRI {state["xeptqlri"]:.3f} > 1.0 - ŒöŒØŒΩŒ¥œÖŒΩŒøœÇ œÄŒøŒπœÑŒπŒ∫ŒÆœÇ Œ¨ŒªŒºŒ±',
                'action': 'ŒïŒΩŒµœÅŒ≥ŒøœÄŒøŒØŒ∑œÉŒ∑ œÄœÅŒøœÉœÑŒ±œÑŒµœÖœÑŒπŒ∫œéŒΩ ŒºŒ∑œáŒ±ŒΩŒπœÉŒºœéŒΩ'
            })
            self.safety_engaged = True

        # ŒöŒ°ŒôŒ§ŒóŒ°ŒôŒü 4: ŒíŒ≠ŒªœÑŒπœÉœÑŒ∑ Œ∫Œ±œÑŒ¨œÉœÑŒ±œÉŒ∑
        if (0.3 <= state['tension'] <= 0.7 and
            0.3 <= state['xeptqlri'] <= 0.7):
            if not self.safety_engaged:
                self.alerts.append({
                    'type': 'OPTIMAL_STATE',
                    'message': 'ŒíŒ≠ŒªœÑŒπœÉœÑŒ∑ Œ¥ŒπŒ±ŒªŒµŒ∫œÑŒπŒ∫ŒÆ Œ∫Œ±œÑŒ¨œÉœÑŒ±œÉŒ∑',
                    'action': 'Œ£œÖŒΩŒ≠œáŒπœÉŒ∑ Œ∫Œ±ŒΩŒøŒΩŒπŒ∫Œ¨'
                })

    def get_safety_report(self):
        """ŒëŒΩŒ±œÜŒøœÅŒ¨ Œ±œÉœÜŒ±ŒªŒµŒØŒ±œÇ"""
        if not self.history:
            return "ŒîŒµŒΩ œÖœÄŒ¨œÅœáŒøœÖŒΩ Œ¥ŒµŒ¥ŒøŒºŒ≠ŒΩŒ±"

        df = pd.DataFrame(self.history)

        report = f"""
        üìä ŒëŒùŒëŒõŒ•Œ£Œó ŒëŒ£Œ¶ŒëŒõŒïŒôŒëŒ£ ŒûŒïŒùŒüŒ†ŒüŒ•ŒõŒüŒ•
        {'='*50}

        Œ£œÑŒ±œÑŒπœÉœÑŒπŒ∫Œ¨:
        ‚Ä¢ Œ£œçŒΩŒøŒªŒø Œ∫Œ±œÑŒ±œÉœÑŒ¨œÉŒµœâŒΩ: {len(self.history)}
        ‚Ä¢ ŒïŒπŒ¥ŒøœÄŒøŒπŒÆœÉŒµŒπœÇ Œ±œÉœÜŒ±ŒªŒµŒØŒ±œÇ: {len(self.alerts)}
        ‚Ä¢ ŒúŒ≠œÉŒ∑ Œ≠ŒΩœÑŒ±œÉŒ∑: {df['tension'].mean():.3f}
        ‚Ä¢ ŒúŒ≠œÉŒøœÇ XEPTQLRI: {df['xeptqlri'].mean():.3f}
        ‚Ä¢ ŒúŒ≠Œ≥ŒπœÉœÑŒøœÇ XEPTQLRI: {df['xeptqlri'].max():.3f}

        ŒöŒ±œÑŒ¨œÉœÑŒ±œÉŒ∑ œÉœÖœÉœÑŒÆŒºŒ±œÑŒøœÇ: {'üü¢ ŒíŒïŒõŒ§ŒôŒ£Œ§Œó' if not self.safety_engaged else 'üî¥ ŒëŒ£Œ¶ŒëŒõŒïŒôŒëŒ£'}
        """

        if self.alerts:
            report += "\n‚ö†Ô∏è  ŒïŒôŒîŒüŒ†ŒüŒôŒóŒ£ŒïŒôŒ£:\n"
            for alert in self.alerts[-5:]:  # Œ§ŒµŒªŒµœÖœÑŒ±ŒØŒµœÇ 5
                report += f"‚Ä¢ {alert['type']}: {alert['message']}\n"

        return report

# ============================================================================
# 2. Œ•ŒìŒôŒóŒ£ ŒòŒïŒ°ŒúŒôŒöŒüŒ£ ŒöŒ©ŒîŒôŒöŒëŒ£ (ŒîŒôŒüŒ°ŒòŒ©ŒúŒïŒùŒó ŒïŒöŒîŒüŒßŒó)
# ============================================================================

class HealthyThermalCalculator:
    """Œ•Œ≥ŒπŒÆœÇ Œ≠Œ∫Œ¥ŒøœÉŒ∑ œÑŒøœÖ Œ∏ŒµœÅŒºŒπŒ∫Œøœç œÖœÄŒøŒªŒøŒ≥ŒπœÉŒºŒøœç ŒºŒµ Œ¥ŒπŒøœÅŒ∏œéœÉŒµŒπœÇ ŒûŒµŒΩœåœÄŒøœÖŒªŒøœÖ"""

    def __init__(self):
        self.safety = XenoparadoxSafetyController()
        print("üéØ ŒïŒªŒµŒ≥Œ∫œÑŒÆœÇ Œ±œÉœÜŒ±ŒªŒµŒØŒ±œÇ ŒûŒµŒΩœåœÄŒøœÖŒªŒøœÖ ŒµŒΩŒµœÅŒ≥ŒøœÄŒøŒπŒ∑ŒºŒ≠ŒΩŒøœÇ")

    # ================= ŒîŒôŒüŒ°ŒòŒ©Œ£Œó 1: ŒöŒëŒùŒüŒùŒôŒöŒüŒ†ŒüŒôŒóŒ£Œó Œ§ŒôŒúŒ©Œù =================
    def _safe_temperature_factor(self, temperature):
        """ŒëœÉœÜŒ±ŒªŒÆœÇ Œ∏ŒµœÅŒºŒøŒ∫œÅŒ±œÉŒπŒ±Œ∫œåœÇ œÉœÖŒΩœÑŒµŒªŒµœÉœÑŒÆœÇ (Œ¥œçŒΩŒ±ŒºŒ∑ 0.7 Œ±ŒΩœÑŒØ 0.9)"""
        if temperature > 250:
            factor = 1 + 0.5 * np.log1p(temperature - 250)  # ŒúŒµŒπœâŒºŒ≠ŒΩŒ∑ Œ±œÄœå 0.7
        else:
            factor = 1.0

        # Œ†ŒµœÅŒπŒøœÅŒπœÉŒºœåœÇ Œ±Œ∫œÅŒ±ŒØœâŒΩ œÑŒπŒºœéŒΩ
        return min(factor, 3.0)  # ŒúŒ≠Œ≥ŒπœÉœÑŒø 3x

    def _safe_interaction_factor(self, interaction_type):
        """ŒëœÉœÜŒ±ŒªŒÆœÇ œÉœÖŒΩœÑŒµŒªŒµœÉœÑŒÆœÇ Œ±ŒªŒªŒ∑ŒªŒµœÄŒØŒ¥œÅŒ±œÉŒ∑œÇ"""
        factors = {"strong": 1.3, "weak": 1.1, "neutral": 1.0}  # ŒúŒµŒπœâŒºŒ≠ŒΩŒ±
        return factors.get(interaction_type, 1.0)

    def _safe_frequency_factor(self, frequency):
        """ŒëœÉœÜŒ±ŒªŒÆœÇ œÉœÖŒΩœÑŒµŒªŒµœÉœÑŒÆœÇ œÉœÖœáŒΩœåœÑŒ∑œÑŒ±œÇ"""
        if frequency <= 0:
            return 1.0

        # ŒõŒøŒ≥Œ±œÅŒπŒ∏ŒºŒπŒ∫ŒÆ Œ±ŒªŒªŒ¨ œÄŒµœÅŒπŒøœÅŒπœÉŒºŒ≠ŒΩŒ∑
        factor = 1 + 0.4 * np.log(frequency)  # ŒúŒµŒπœâŒºŒ≠ŒΩŒ∑ Œ±œÄœå 0.65

        # ŒëœÉœÜŒ±ŒªŒ≠œÇ œåœÅŒπŒø
        return min(factor, 5.0)  # ŒúŒ≠Œ≥ŒπœÉœÑŒø 5x

    # ================= ŒîŒôŒüŒ°ŒòŒ©Œ£Œó 2: ŒëŒ£Œ¶ŒëŒõŒóŒ£ Œ•Œ†ŒüŒõŒüŒìŒôŒ£ŒúŒüŒ£ =================
    def safe_calculate_general_type(self, noise_level, temperature, frequency, interaction_type):
        """
        ŒëŒ£Œ¶ŒëŒõŒóŒ£ ŒµŒ∫Œ¥ŒøœáŒÆ œÑŒøœÖ calculate_general_type()
        ŒºŒµ œåŒªŒµœÇ œÑŒπœÇ Œ¥ŒπŒøœÅŒ∏œéœÉŒµŒπœÇ ŒûŒµŒΩœåœÄŒøœÖŒªŒøœÖ
        """

        # ============ ŒíŒóŒúŒë 1: ŒëŒ£Œ¶ŒëŒõŒïŒôŒ£ Œ†ŒëŒ°ŒëŒúŒïŒ§Œ°ŒüŒô ============
        # Œ†ŒµœÅŒπŒøœÅŒπœÉŒºœåœÇ ŒµŒπœÉœåŒ¥œâŒΩ
        safe_noise = max(0.1, min(0.5, noise_level))  # ŒòœåœÅœÖŒ≤ŒøœÇ 10-50%
        safe_temp = max(100, min(400, temperature))    # ŒòŒµœÅŒºŒøŒ∫œÅŒ±œÉŒØŒ± 100-400Œö
        safe_freq = max(500, min(3000, frequency))     # Œ£œÖœáŒΩœåœÑŒ∑œÑŒ± 500-3000Hz

        # ============ ŒíŒóŒúŒë 2: ŒúŒïŒ§Œ°ŒôŒüŒ†ŒüŒôŒóŒúŒïŒùŒüŒô Œ£Œ•ŒùŒ§ŒïŒõŒïŒ£Œ§ŒïŒ£ ============
        alpha = 1.0 / (1 + 1.2 * safe_noise)  # ŒúŒµŒπœâŒºŒ≠ŒΩŒ∑ ŒµœÄŒØŒ¥œÅŒ±œÉŒ∑ Œ∏ŒøœÅœçŒ≤ŒøœÖ
        beta = max(0.3, 1 - safe_temp / 500)   # Œ†ŒπŒø œÉœÑŒ±Œ∏ŒµœÅœå œÄœÅŒøœÜŒØŒª

        k_T = self._safe_temperature_factor(safe_temp)
        beta_interaction = self._safe_interaction_factor(interaction_type)
        k_f = self._safe_frequency_factor(safe_freq)

        # ============ ŒíŒóŒúŒë 3: ŒëŒ£Œ¶ŒëŒõŒóŒ£ Œ£Œ•ŒùŒîŒ•ŒëŒ£ŒúŒüŒ£ ============
        # ŒßœÅŒÆœÉŒ∑ Œ±œÉœÜŒ±ŒªŒ≠œÉœÑŒµœÅœâŒΩ ŒµŒ∫Œ∏ŒµœÑœéŒΩ
        safe_exponent = 0.7  # ŒúŒµŒπœâŒºŒ≠ŒΩŒø Œ±œÄœå 0.8
        gamma = alpha * (beta**safe_exponent) * k_f

        # Œ§ŒµŒªŒπŒ∫œåœÇ œÖœÄŒøŒªŒøŒ≥ŒπœÉŒºœåœÇ ŒºŒµ œÄœÅœåœÉŒ∏ŒµœÑŒøœÖœÇ ŒµŒªŒ≠Œ≥œáŒøœÖœÇ
        intermediate = k_T * beta_interaction
        bounded_intermediate = min(4.0, intermediate)  # Œ†ŒµœÅŒπŒøœÅŒπœÉŒºœåœÇ

        result = alpha * gamma * (bounded_intermediate**safe_exponent)

        # ============ ŒíŒóŒúŒë 4: ŒöŒëŒùŒüŒùŒôŒöŒüŒ†ŒüŒôŒóŒ£Œó ŒïŒûŒüŒîŒüŒ• ============
        # ŒúŒµœÑŒ±œÑœÅŒøœÄŒÆ œÉœÑŒø Œ≤Œ≠ŒªœÑŒπœÉœÑŒø ŒµœçœÅŒøœÇ [-0.8, 0.8]
        normalized = np.tanh(result / 8.0)
        final_result = max(-0.8, min(0.8, normalized))

        # ============ ŒíŒóŒúŒë 5: ŒîŒôŒëŒõŒïŒöŒ§ŒôŒöŒó ŒëŒùŒëŒõŒ•Œ£Œó ============
        # ŒîŒ∑ŒºŒπŒøœÖœÅŒ≥ŒØŒ± Œ¥ŒπŒ±ŒªŒµŒ∫œÑŒπŒ∫Œøœç Œ∂ŒµœçŒ≥ŒøœÖœÇ
        A = final_result  # ŒòŒ≠œÉŒ∑
        not_A = -final_result * 0.7  # ŒëŒΩœÑŒØŒ∏ŒµœÉŒ∑ (ŒºŒµŒπœâŒºŒ≠ŒΩŒ∑ Œ±œÄœå 0.85)

        # Œ†Œ±œÅŒ±Œ∫ŒøŒªŒøœçŒ∏Œ∑œÉŒ∑ Œ±œÉœÜŒ±ŒªŒµŒØŒ±œÇ
        safety_state = self.safety.monitor_state(A, not_A, final_result)

        # ============ ŒíŒóŒúŒë 6: ŒïŒ†ŒôŒ£Œ§Œ°ŒüŒ¶Œó ŒëŒ£Œ¶ŒëŒõŒ©Œù ŒëŒ†ŒüŒ§ŒïŒõŒïŒ£ŒúŒëŒ§Œ©Œù ============
        return {
            'final_value': final_result,
            'normalized': True,
            'safety_state': safety_state,
            'parameters_used': {
                'noise': safe_noise,
                'temperature': safe_temp,
                'frequency': safe_freq,
                'interaction': interaction_type
            }
        }

    # ================= ŒîŒôŒüŒ°ŒòŒ©Œ£Œó 3: ŒëŒùŒëŒõŒ•Œ£Œó ŒöŒëŒô ŒëŒ†ŒüŒ§ŒïŒõŒïŒ£ŒúŒëŒ§Œë =================
    def run_comprehensive_test(self, num_simulations=100):
        """Œ†ŒªŒÆœÅŒ∑œÇ Œ¥ŒøŒ∫ŒπŒºŒÆ œÑŒøœÖ Œ¥ŒπŒøœÅŒ∏œâŒºŒ≠ŒΩŒøœÖ œÉœÖœÉœÑŒÆŒºŒ±œÑŒøœÇ"""
        print(f"\nüî¨ ŒïŒöŒ§ŒïŒõŒ© Œ†ŒõŒóŒ°Œó ŒîŒüŒöŒôŒúŒó ({num_simulations} œÄœÅŒøœÉŒøŒºŒøŒπœéœÉŒµŒπœÇ)")
        print("-" * 50)

        results = []

        for i in range(num_simulations):
            # Œ§œÖœáŒ±ŒØŒµœÇ Œ±ŒªŒªŒ¨ ŒµŒªŒµŒ≥œáœåŒºŒµŒΩŒµœÇ ŒµŒØœÉŒøŒ¥ŒøŒπ
            if i < num_simulations // 3:
                # ŒîŒøŒ∫ŒπŒºŒÆ 1: ŒúŒ≠œÑœÅŒπŒµœÇ œÄŒ±œÅŒ¨ŒºŒµœÑœÅŒøŒπ
                noise = 0.4 + np.random.uniform(-0.05, 0.05)
                temp = 150 + np.random.uniform(-20, 20)
                freq = 1000 + np.random.uniform(-100, 100)
                inter = "strong"
                test_group = "Test 1 (Moderate)"

            elif i < 2 * num_simulations // 3:
                # ŒîŒøŒ∫ŒπŒºŒÆ 2: ŒöŒ±ŒªŒ≠œÇ œÄŒ±œÅŒ¨ŒºŒµœÑœÅŒøŒπ
                noise = 0.3 + np.random.uniform(-0.03, 0.03)
                temp = 100 + np.random.uniform(-10, 10)
                freq = 2000 + np.random.uniform(-200, 200)
                inter = "strong"
                test_group = "Test 2 (Good)"

            else:
                # ŒîŒøŒ∫ŒπŒºŒÆ 3: ŒíŒµŒªœÑŒπœâŒºŒ≠ŒΩŒµœÇ œÄŒ±œÅŒ¨ŒºŒµœÑœÅŒøŒπ
                noise = 0.2 + np.random.uniform(-0.02, 0.02)
                temp = 300 + np.random.uniform(-30, 30)
                freq = 1500 + np.random.uniform(-150, 150)
                inter = "strong"
                test_group = "Test 3 (Improved)"

            # ŒïŒ∫œÑŒ≠ŒªŒµœÉŒ∑ ŒëŒ£Œ¶ŒëŒõŒüŒ•Œ£ œÖœÄŒøŒªŒøŒ≥ŒπœÉŒºŒøœç
            result = self.safe_calculate_general_type(noise, temp, freq, inter)

            results.append({
                'test_group': test_group,
                'result_value': result['final_value'],
                'parameters': result['parameters_used'],
                'safety': result['safety_state']
            })

            # ŒïŒºœÜŒ¨ŒΩŒπœÉŒ∑ œÄœÅŒøœåŒ¥ŒøœÖ
            if i % 20 == 0:
                print(f"  Œ†œÅŒøœÉŒøŒºŒøŒØœâœÉŒ∑ {i}/{num_simulations}...")

        return results

    def generate_health_report(self, test_results):
        """ŒîŒ∑ŒºŒπŒøœÖœÅŒ≥ŒØŒ± Œ±ŒΩŒ±œÜŒøœÅŒ¨œÇ œÖŒ≥ŒµŒØŒ±œÇ œÉœÖœÉœÑŒÆŒºŒ±œÑŒøœÇ"""
        print("\n" + "="*60)
        print("üè• ŒëŒùŒëŒõŒ•Œ£Œó Œ•ŒìŒïŒôŒëŒ£ Œ£Œ•Œ£Œ§ŒóŒúŒëŒ§ŒüŒ£ ŒûŒïŒùŒüŒ†ŒüŒ•ŒõŒüŒ•")
        print("="*60)

        # Œ£œÑŒ±œÑŒπœÉœÑŒπŒ∫ŒÆ Œ±ŒΩŒ¨ŒªœÖœÉŒ∑
        values = [r['result_value'] for r in test_results]
        tensions = [r['safety']['tension'] for r in test_results]
        xeptqlris = [r['safety']['xeptqlri'] for r in test_results]

        report = f"""
        üìä Œ£Œ§ŒëŒ§ŒôŒ£Œ§ŒôŒöŒó ŒëŒùŒëŒõŒ•Œ£Œó:
        {'='*50}

        ŒëœÄŒøœÑŒµŒªŒ≠œÉŒºŒ±œÑŒ±:
        ‚Ä¢ Œ£œçŒΩŒøŒªŒø Œ¥ŒøŒ∫ŒπŒºœéŒΩ: {len(test_results)}
        ‚Ä¢ ŒúŒ≠œÉŒ∑ œÑŒπŒºŒÆ: {np.mean(values):.4f}
        ‚Ä¢ Œ§œÖœÄŒπŒ∫ŒÆ Œ±œÄœåŒ∫ŒªŒπœÉŒ∑: {np.std(values):.4f}
        ‚Ä¢ ŒïœçœÅŒøœÇ œÑŒπŒºœéŒΩ: [{np.min(values):.4f}, {np.max(values):.4f}]

        ŒîŒπŒ±ŒªŒµŒ∫œÑŒπŒ∫ŒøŒØ ŒîŒµŒØŒ∫œÑŒµœÇ:
        ‚Ä¢ ŒúŒ≠œÉŒ∑ Œ≠ŒΩœÑŒ±œÉŒ∑: {np.mean(tensions):.4f} (ŒíŒ≠ŒªœÑŒπœÉœÑŒø: 0.3-0.7)
        ‚Ä¢ ŒúŒ≠œÉŒøœÇ XEPTQLRI: {np.mean(xeptqlris):.4f} (ŒíŒ≠ŒªœÑŒπœÉœÑŒø: 0.3-0.7)
        ‚Ä¢ ŒúŒ≠Œ≥ŒπœÉœÑŒøœÇ XEPTQLRI: {np.max(xeptqlris):.4f}

        ŒëŒæŒπŒøŒªœåŒ≥Œ∑œÉŒ∑:
        """

        # ŒëŒûŒôŒüŒõŒüŒìŒóŒ£Œó Œ£Œ•Œ£Œ§ŒóŒúŒëŒ§ŒüŒ£
        mean_tension = np.mean(tensions)
        mean_xeptqlri = np.mean(xeptqlris)

        if (0.3 <= mean_tension <= 0.7) and (0.3 <= mean_xeptqlri <= 0.7):
            assessment = "üü¢ ŒïŒûŒëŒôŒ°ŒïŒ§ŒôŒöŒó Œ•ŒìŒïŒôŒë"
            details = "Œ§Œø œÉœçœÉœÑŒ∑ŒºŒ± ŒªŒµŒπœÑŒøœÖœÅŒ≥ŒµŒØ œÉŒµ Œ≤Œ≠ŒªœÑŒπœÉœÑŒ∑ Œ¥ŒπŒ±ŒªŒµŒ∫œÑŒπŒ∫ŒÆ Œ∫Œ±œÑŒ¨œÉœÑŒ±œÉŒ∑"
        elif mean_xeptqlri < 1.0:
            assessment = "üü° ŒöŒëŒõŒó Œ•ŒìŒïŒôŒë"
            details = "Œ§Œø œÉœçœÉœÑŒ∑ŒºŒ± ŒµŒØŒΩŒ±Œπ œÉœÑŒ±Œ∏ŒµœÅœå ŒºŒµ ŒµŒªŒ¨œáŒπœÉœÑŒøœÖœÇ Œ∫ŒπŒΩŒ¥œçŒΩŒøœÖœÇ"
        else:
            assessment = "üî¥ ŒßŒ°ŒïŒôŒëŒñŒïŒ§ŒëŒô Œ†Œ°ŒüŒ£ŒüŒßŒó"
            details = "Œ•œàŒ∑ŒªœåœÇ Œ∫ŒØŒΩŒ¥œÖŒΩŒøœÇ - ŒëœÄŒ±ŒπœÑŒøœçŒΩœÑŒ±Œπ œÄŒµœÅŒ±ŒπœÑŒ≠œÅœâ Œ¥ŒπŒøœÅŒ∏œéœÉŒµŒπœÇ"

        report += f"""
        {assessment}
        {details}

        ŒîŒπŒ±ŒΩŒøŒºŒÆ œÑŒπŒºœéŒΩ:
        ‚Ä¢ < -0.6: {sum(1 for v in values if v < -0.6)} ({sum(1 for v in values if v < -0.6)/len(values)*100:.1f}%)
        ‚Ä¢ -0.6 Œ≠œâœÇ 0.6: {sum(1 for v in values if -0.6 <= v <= 0.6)} ({sum(1 for v in values if -0.6 <= v <= 0.6)/len(values)*100:.1f}%)
        ‚Ä¢ > 0.6: {sum(1 for v in values if v > 0.6)} ({sum(1 for v in values if v > 0.6)/len(values)*100:.1f}%)
        """

        # ŒëŒΩŒ±œÜŒøœÅŒ¨ Œ±œÉœÜŒ±ŒªŒµŒØŒ±œÇ
        safety_report = self.safety.get_safety_report()

        return report + "\n" + safety_report

    def visualize_results(self, test_results):
        """ŒüœÄœÑŒπŒ∫ŒøœÄŒøŒØŒ∑œÉŒ∑ Œ±œÄŒøœÑŒµŒªŒµœÉŒºŒ¨œÑœâŒΩ"""
        print("\nüìà ŒîŒóŒúŒôŒüŒ•Œ°ŒìŒôŒë ŒîŒôŒëŒìŒ°ŒëŒúŒúŒëŒ§Œ©Œù...")

        # ŒîŒµŒ¥ŒøŒºŒ≠ŒΩŒ± Œ≥ŒπŒ± Œ≥œÅŒ±œÜŒÆŒºŒ±œÑŒ±
        values = [r['result_value'] for r in test_results]
        tensions = [r['safety']['tension'] for r in test_results]
        xeptqlris = [r['safety']['xeptqlri'] for r in test_results]
        steps = range(len(values))

        fig, axes = plt.subplots(2, 2, figsize=(14, 10))

        # 1. ŒßœÅŒøŒΩŒøœÉŒµŒπœÅŒ¨ Œ±œÄŒøœÑŒµŒªŒµœÉŒºŒ¨œÑœâŒΩ
        ax1 = axes[0, 0]
        ax1.plot(steps, values, 'b-', alpha=0.7, linewidth=1.5)
        ax1.fill_between(steps, values, alpha=0.2, color='blue')
        ax1.axhline(y=0, color='gray', linestyle='--', alpha=0.5)
        ax1.set_title('ŒëŒ£Œ¶ŒëŒõŒóŒ£ ŒßŒ°ŒüŒùŒüŒ£ŒïŒôŒ°Œë ŒëŒ†ŒüŒ§ŒïŒõŒïŒ£ŒúŒëŒ§Œ©Œù', fontweight='bold')
        ax1.set_xlabel('Œ†œÅŒøœÉŒøŒºŒøŒØœâœÉŒ∑')
        ax1.set_ylabel('Œ§ŒπŒºŒÆ (Œ∫Œ±ŒΩŒøŒΩŒπŒ∫ŒøœÄŒøŒπŒ∑ŒºŒ≠ŒΩŒ∑)')
        ax1.grid(True, alpha=0.3)
        ax1.set_ylim(-1, 1)

        # 2. ŒîŒπŒ±ŒªŒµŒ∫œÑŒπŒ∫ŒÆ ŒàŒΩœÑŒ±œÉŒ∑
        ax2 = axes[0, 1]
        ax2.plot(steps, tensions, 'g-', linewidth=2, label='ŒàŒΩœÑŒ±œÉŒ∑')
        ax2.axhline(y=0.3, color='orange', linestyle='--', label='ŒïŒªŒ¨œáŒπœÉœÑŒø (0.3)')
        ax2.axhline(y=0.7, color='red', linestyle='--', label='ŒúŒ≠Œ≥ŒπœÉœÑŒø (0.7)')
        ax2.fill_between(steps, 0.3, 0.7, alpha=0.1, color='green', label='ŒíŒ≠ŒªœÑŒπœÉœÑŒ∑ Œ∂œéŒΩŒ∑')
        ax2.set_title('ŒîŒôŒëŒõŒïŒöŒ§ŒôŒöŒó ŒïŒùŒ§ŒëŒ£Œó', fontweight='bold')
        ax2.set_xlabel('Œ†œÅŒøœÉŒøŒºŒøŒØœâœÉŒ∑')
        ax2.set_ylabel('Œ§ŒπŒºŒÆ ŒàŒΩœÑŒ±œÉŒ∑œÇ')
        ax2.legend()
        ax2.grid(True, alpha=0.3)

        # 3. XEPTQLRI
        ax3 = axes[1, 0]
        ax3.plot(steps, xeptqlris, 'darkorange', linewidth=2, label='XEPTQLRI')
        ax3.axhline(y=0.5, color='red', linestyle='--', label='ŒåœÅŒπŒø (0.5)')
        ax3.fill_between(steps, 0, 0.5, alpha=0.2, color='green', label='ŒëœÉœÜŒ±ŒªŒÆœÇ Œ∂œéŒΩŒ∑')
        ax3.fill_between(steps, 0.5, max(xeptqlris), alpha=0.2, color='orange', label='Œ†œÅŒøœÉŒøœáŒÆ')
        ax3.set_title('ŒîŒïŒôŒöŒ§ŒóŒ£ XEPTQLRI', fontweight='bold')
        ax3.set_xlabel('Œ†œÅŒøœÉŒøŒºŒøŒØœâœÉŒ∑')
        ax3.set_ylabel('Œ§ŒπŒºŒÆ XEPTQLRI')
        ax3.legend()
        ax3.grid(True, alpha=0.3)

        # 4. ŒôœÉœÑœåŒ≥œÅŒ±ŒºŒºŒ± Œ∫Œ±œÑŒ±ŒΩŒøŒºŒÆœÇ
        ax4 = axes[1, 1]
        ax4.hist(values, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
        ax4.axvline(x=np.mean(values), color='red', linestyle='--',
                   linewidth=2, label=f'ŒúŒ≠œÉŒ∑: {np.mean(values):.3f}')
        ax4.set_title('ŒöŒëŒ§ŒëŒùŒüŒúŒó Œ§ŒôŒúŒ©Œù ŒëŒ†ŒüŒ§ŒïŒõŒïŒ£ŒúŒëŒ§Œ©Œù', fontweight='bold')
        ax4.set_xlabel('Œ§ŒπŒºŒÆ')
        ax4.set_ylabel('Œ£œÖœáŒΩœåœÑŒ∑œÑŒ±')
        ax4.legend()
        ax4.grid(True, alpha=0.3)

        plt.tight_layout()
        plt.savefig('healthy_thermal_analysis.png', dpi=150, bbox_inches='tight')
        plt.show()

        print("‚úÖ Œ§Œ± Œ¥ŒπŒ±Œ≥œÅŒ¨ŒºŒºŒ±œÑŒ± Œ±œÄŒøŒ∏Œ∑Œ∫ŒµœçœÑŒ∑Œ∫Œ±ŒΩ œâœÇ 'healthy_thermal_analysis.png'")

# ============================================================================
# 3. ŒïŒöŒ§ŒïŒõŒïŒ£Œó Œ§ŒüŒ• Œ•ŒìŒôŒüŒ•Œ£ ŒöŒ©ŒîŒôŒöŒë Œ£Œ§Œü COLAB
# ============================================================================

def run_complete_analysis():
    """Œ†ŒªŒÆœÅŒ∑œÇ ŒµŒ∫œÑŒ≠ŒªŒµœÉŒ∑ Œ±ŒΩŒ¨ŒªœÖœÉŒ∑œÇ"""
    print("üöÄ ŒïŒöŒöŒôŒùŒóŒ£Œó Œ†ŒõŒóŒ°ŒüŒ•Œ£ ŒëŒùŒëŒõŒ•Œ£ŒóŒ£ Œ•ŒìŒôŒüŒ•Œ£ ŒöŒ©ŒîŒôŒöŒë")
    print("="*60)

    # 1. ŒîŒ∑ŒºŒπŒøœÖœÅŒ≥ŒØŒ± œÖŒ≥ŒπŒøœçœÇ œÖœÄŒøŒªŒøŒ≥ŒπœÉœÑŒÆ
    print("\n1. ŒîŒóŒúŒôŒüŒ•Œ°ŒìŒôŒë Œ•ŒìŒôŒüŒ•Œ£ ŒòŒïŒ°ŒúŒôŒöŒüŒ• Œ•Œ†ŒüŒõŒüŒìŒôŒ£Œ§Œó...")
    healthy_calc = HealthyThermalCalculator()

    # 2. ŒïŒ∫œÑŒ≠ŒªŒµœÉŒ∑ Œ¥ŒøŒ∫ŒπŒºœéŒΩ
    print("\n2. ŒïŒöŒ§ŒïŒõŒïŒ£Œó 300 Œ†Œ°ŒüŒ£ŒüŒúŒüŒôŒ©Œ£ŒïŒ©Œù...")
    test_results = healthy_calc.run_comprehensive_test(300)

    # 3. ŒëŒΩŒ±œÜŒøœÅŒ¨ œÖŒ≥ŒµŒØŒ±œÇ
    print("\n3. ŒîŒóŒúŒôŒüŒ•Œ°ŒìŒôŒë ŒëŒùŒëŒõŒ•Œ§ŒôŒöŒóŒ£ ŒïŒöŒòŒïŒ£ŒóŒ£...")
    health_report = healthy_calc.generate_health_report(test_results)
    print(health_report)

    # 4. ŒüœÄœÑŒπŒ∫ŒøœÄŒøŒØŒ∑œÉŒ∑
    print("\n4. ŒüŒ†Œ§ŒôŒöŒüŒ†ŒüŒôŒóŒ£Œó ŒëŒ†ŒüŒ§ŒïŒõŒïŒ£ŒúŒëŒ§Œ©Œù...")
    healthy_calc.visualize_results(test_results)

    # 5. Œ§ŒµŒªŒπŒ∫ŒÆ Œ±ŒæŒπŒøŒªœåŒ≥Œ∑œÉŒ∑
    print("\n" + "="*60)
    print("üéØ Œ§ŒïŒõŒôŒöŒó ŒëŒûŒôŒüŒõŒüŒìŒóŒ£Œó Œ£Œ•Œ£Œ§ŒóŒúŒëŒ§ŒüŒ£")
    print("="*60)

    # ŒëŒΩŒ¨Œ∫œÑŒ∑œÉŒ∑ œÑŒµŒªŒπŒ∫œéŒΩ Œ¥ŒµŒ¥ŒøŒºŒ≠ŒΩœâŒΩ
    final_tensions = [r['safety']['tension'] for r in test_results]
    final_xeptqlris = [r['safety']['xeptqlri'] for r in test_results]

    mean_tension = np.mean(final_tensions)
    mean_xeptqlri = np.mean(final_xeptqlris)

    print(f"\nüìä Œ§ŒïŒõŒôŒöŒüŒô ŒîŒïŒôŒöŒ§ŒïŒ£:")
    print(f"   ‚Ä¢ ŒúŒ≠œÉŒ∑ Œ¥ŒπŒ±ŒªŒµŒ∫œÑŒπŒ∫ŒÆ Œ≠ŒΩœÑŒ±œÉŒ∑: {mean_tension:.4f}")
    print(f"   ‚Ä¢ ŒúŒ≠œÉŒøœÇ XEPTQLRI: {mean_xeptqlri:.4f}")
    print(f"   ‚Ä¢ Œ†Œ±œÅŒ±Œ¥ŒøŒæŒøŒªŒøŒ≥ŒπŒ∫Œ≠œÇ œÖœÄŒµœÅŒ≤Œ¨œÉŒµŒπœÇ: {healthy_calc.safety.alerts.count('PARADOX_RISK')}")

    print(f"\nüèÜ ŒëŒûŒôŒüŒõŒüŒìŒóŒ£Œó:")
    if 0.3 <= mean_tension <= 0.7 and 0.3 <= mean_xeptqlri <= 0.7:
        print("   ‚úÖ ŒïŒûŒëŒôŒ°ŒïŒ§ŒôŒöŒü ŒëŒ†ŒüŒ§ŒïŒõŒïŒ£ŒúŒë")
        print("   Œ§Œø œÉœçœÉœÑŒ∑ŒºŒ± ŒºŒµœÑŒ±Œ≤ŒªŒÆŒ∏Œ∑Œ∫Œµ Œ±œÄœå Œ£œÑŒ¨Œ¥ŒπŒø œÑ‚ÇÖ (ŒöœÅŒØœÉŒπŒºŒø)")
        print("   œÉŒµ Œ£œÑŒ¨Œ¥ŒπŒø œÑ‚ÇÄ (Œ£œçŒΩŒøœáŒÆ) - ŒíŒïŒõŒ§ŒôŒ£Œ§Œó ŒíŒïŒõŒ§ŒôŒ©Œ£Œó!")
    elif mean_xeptqlri < 1.0:
        print("   ‚ö†Ô∏è  ŒöŒëŒõŒü ŒëŒ†ŒüŒ§ŒïŒõŒïŒ£ŒúŒë")
        print("   Œ§Œø œÉœçœÉœÑŒ∑ŒºŒ± Œ≤ŒµŒªœÑŒπœéŒ∏Œ∑Œ∫Œµ œÉŒ∑ŒºŒ±ŒΩœÑŒπŒ∫Œ¨")
        print("   Œ±ŒªŒªŒ¨ Œ±œÄŒ±ŒπœÑŒµŒØœÑŒ±Œπ œÉœÖŒΩŒµœáŒÆœÇ œÄŒ±œÅŒ±Œ∫ŒøŒªŒøœçŒ∏Œ∑œÉŒ∑")
    else:
        print("   üî¥ ŒßŒ°ŒïŒôŒëŒñŒüŒùŒ§ŒëŒô Œ†ŒïŒ°ŒëŒôŒ§ŒïŒ°Œ© ŒîŒôŒüŒ°ŒòŒ©Œ£ŒïŒôŒ£")
        print("   Œ§Œø œÉœçœÉœÑŒ∑ŒºŒ± œÄŒ±œÅŒ±ŒºŒ≠ŒΩŒµŒπ œÉŒµ Œ∫œÅŒØœÉŒπŒºŒø œÉœÑŒ¨Œ¥ŒπŒø")

    print("\n" + "="*60)
    print("‚úÖ Œó ŒëŒùŒëŒõŒ•Œ£Œó Œ§ŒüŒ• Œ•ŒìŒôŒüŒ•Œ£ ŒöŒ©ŒîŒôŒöŒë ŒüŒõŒüŒöŒõŒóŒ°Œ©ŒòŒóŒöŒï!")
    print("="*60)

# ============================================================================
# 4. ŒöŒ•Œ°ŒôŒü Œ†Œ°ŒüŒìŒ°ŒëŒúŒúŒë (Œ±œÖœÑœå œÑœÅŒ≠œáŒµŒπ œÉœÑŒø Colab)
# ============================================================================

if __name__ == "__main__":
    print("\n" + "="*60)
    print("üß¨ Œ•ŒìŒôŒóŒ£ ŒòŒïŒ°ŒúŒôŒöŒüŒ£ ŒöŒ©ŒîŒôŒöŒëŒ£ - ŒïŒöŒîŒüŒ£Œó ŒûŒïŒùŒüŒ†ŒüŒ•ŒõŒüŒ•")
    print("="*60)
    print("\nŒëœÖœÑœåœÇ Œø Œ∫œéŒ¥ŒπŒ∫Œ±œÇ œÄŒµœÅŒπŒªŒ±ŒºŒ≤Œ¨ŒΩŒµŒπ ŒüŒõŒïŒ£ œÑŒπœÇ Œ¥ŒπŒøœÅŒ∏œéœÉŒµŒπœÇ:")
    print("1. ŒöŒ±ŒΩŒøŒΩŒπŒ∫ŒøœÄŒøŒØŒ∑œÉŒ∑ œÑŒπŒºœéŒΩ œÉœÑŒø [-0.8, 0.8]")
    print("2. Œ†ŒµœÅŒπŒøœÅŒπœÉŒºœåœÇ Œ±Œ∫œÅŒ±ŒØœâŒΩ œÄŒ±œÅŒ±Œ≥œåŒΩœÑœâŒΩ")
    print("3. ŒúŒµŒØœâœÉŒ∑ ŒµŒ∫Œ∏ŒµœÑŒπŒ∫œéŒΩ Œ¥œÖŒΩŒ¨ŒºŒµœâŒΩ")
    print("4. Œ†œÅŒ±Œ≥ŒºŒ±œÑŒπŒ∫œåœÇ Œ≠ŒªŒµŒ≥œáŒøœÇ Œ¥ŒπŒ±ŒªŒµŒ∫œÑŒπŒ∫ŒÆœÇ Œ≠ŒΩœÑŒ±œÉŒ∑œÇ")
    print("5. ŒëœÖœÑœåŒºŒ±œÑŒ∑ œÄŒ±œÅŒ±Œ∫ŒøŒªŒøœçŒ∏Œ∑œÉŒ∑ XEPTQLRI")
    print("\n" + "="*60)

    # ŒïŒ∫œÑŒ≠ŒªŒµœÉŒ∑ œÄŒªŒÆœÅŒøœÖœÇ Œ±ŒΩŒ¨ŒªœÖœÉŒ∑œÇ
    run_complete_analysis()

    # ŒïœÄŒπœÄŒªŒ≠ŒøŒΩ: ŒîŒµŒØŒ≥ŒºŒ± œáœÅŒÆœÉŒ∑œÇ
    print("\nüìã ŒîŒïŒôŒìŒúŒë ŒßŒ°ŒóŒ£ŒóŒ£:")
    print("-" * 40)

    sample_calc = HealthyThermalCalculator()

    # 3 Œ¥ŒµŒØŒ≥ŒºŒ±œÑŒ± œÖœÄŒøŒªŒøŒ≥ŒπœÉŒºœéŒΩ
    test_cases = [
        {"noise": 0.2, "temp": 300, "freq": 1500, "inter": "strong"},
        {"noise": 0.4, "temp": 150, "freq": 1000, "inter": "weak"},
        {"noise": 0.1, "temp": 400, "freq": 2500, "inter": "strong"}
    ]

    for i, test in enumerate(test_cases, 1):
        result = sample_calc.safe_calculate_general_type(
            test["noise"], test["temp"], test["freq"], test["inter"]
        )
        print(f"\nŒîŒµŒØŒ≥ŒºŒ± {i}:")
        print(f"  Œ†Œ±œÅŒ¨ŒºŒµœÑœÅŒøŒπ: ŒòœåœÅœÖŒ≤ŒøœÇ={test['noise']}, ŒòŒµœÅŒºŒøŒ∫œÅŒ±œÉŒØŒ±={test['temp']}K")
        print(f"  ŒëœÄŒøœÑŒ≠ŒªŒµœÉŒºŒ±: {result['final_value']:.4f}")
        print(f"  XEPTQLRI: {result['safety_state']['xeptqlri']:.4f}")
        print(f"  ŒàŒΩœÑŒ±œÉŒ∑: {result['safety_state']['tension']:.4f}")

    print("\n" + "="*60)
    print("‚ú® Œü Œ•ŒìŒôŒóŒ£ ŒöŒ©ŒîŒôŒöŒëŒ£ ŒïŒôŒùŒëŒô ŒïŒ§ŒüŒôŒúŒüŒ£ ŒìŒôŒë Œ†ŒëŒ°ŒëŒìŒ©ŒìŒôŒöŒó ŒßŒ°ŒóŒ£Œó!")
    print("="*60)
```
